```python
import pandas as pd

# Define the file path of your CSV file on Google Drive
csv_file_path = 'https://raw.githubusercontent.com/capitalone/DataProfiler/main/dataprofiler/tests/data/csv/flight_delays.csv'

# Read the CSV file using pandas
df = pd.read_csv(csv_file_path)
df.head()
```

```python
#import dataprofiler as dp
import json
from dataprofiler import Data, Profiler
# profile the data\
#data = Data(df)
profile = Profiler(df, profiler_type='structured')
profile.report(report_options={"output_format":"pretty"})
```

```python
print(type(profile))
#print(profile.row_count)
print(profile.row_has_null_count)
dir(profile)
print(type(profile.report()))
dir(profile.report())
print(profile.report().keys)
list(profile._col_name_to_idx.keys())
print(profile.times)
#print(profile.__dir__())
#print(profile.report()["global_stats"])
print(profile.report()["global_stats"]['samples_used'])
print(profile.report()["global_stats"]['column_count'])
print(profile.report()["global_stats"]['row_count'])
print(profile.report()["global_stats"]['row_has_null_ratio'])
print(profile.report()["global_stats"]['profile_schema'])
for col in profile.report()["global_stats"]['profile_schema']:
    print(col)
for col in profile.report()['data_stats']:
    print(col)
    print(col['column_name'])
    print(col['data_type'])
    print(col['order'])
    for sam in col['samples']:
        print(sam)
        print('stats')
        print(type(col['statistics']))
        print((col['statistics']['min']))
        print((col['statistics']['max']))
        print((col['statistics']['mode']))
    #for stat in col['statistics']:
    #    print(type(stat))
```
